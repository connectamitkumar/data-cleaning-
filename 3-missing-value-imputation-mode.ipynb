```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/house-price-prediction-dataset-81-columns/house_dataset.csv")
```

```python
df.head()
```

```python
cat_var=df.select_dtypes(include='object')
cat_var
```

```python
cat_var.shape
```

```python
miss_value_percentage=cat_var.isnull().mean()*100
```

```python
miss_value_percentage
```

```python
drop_columns=miss_value_percentage[miss_value_percentage>20].keys()
drop_columns
```

```python

cat_var.drop(columns=drop_columns,axis=1,inplace=True)
```

```python
cat_var.shape
```

```python
isnull_per=cat_var.isnull().mean()*100
miss_vars=isnull_per[isnull_per>0].keys()
miss_vars
```

```python
cat_var['MasVnrType'].fillna("missing")
```

```python
cat_var['MasVnrType'].mode()
```

```python
cat_var['MasVnrType'].value_counts()
```

```python
cat_var['MasVnrType'].fillna(cat_var['MasVnrType'].mode()[0])
```

```python
for var in miss_vars:
    cat_var[var].fillna(cat_var[var].mode()[0],inplace=True)
    print(var,"=",cat_var[var].mode()[0])
```
